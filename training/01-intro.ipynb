# grouped calculations

```python
# VERİ SETİMİZİN ILK 5 SATIRI

df.head()
```

```python
# NUMPY İLE VERİ SETİMİZİN ORTALAMA HESABI

import numpy as np
df.groupby('year')['lifeExp'].agg(np.mean)

#df.groupby('year')['lifeExp'].mean()
```

```python
# groupby İLE VERİMİZİ YIL VE ÜLKE BAZINDA AYIRMA İŞLEMİ

df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
```

```python
# RESET INDEX SIRASI VERMEYE YARIYOR

df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean().reset_index()
```

```python
# syntax AÇISINDAN YAZIM ŞEKLİ BOYLE DE OLABILIR

df\
    .groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']]\
    .mean()
```

```python
# syntax AÇISINDAN YAZIM ŞEKLİ BOYLE DE OLABILIR

(df
    .groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']]
    .mean()
    .reset_index()
)
```
