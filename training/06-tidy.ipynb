```python
# DİĞER VERİ DOSYAMIZ
# JOURNAL OF STATİSTİCAL SOFTWARE TİDY DATASET

billboard = pd.read_csv('../data/billboard.csv')
```

```python
billboard.head()
```

```python
# melt()
# id_vars = İSTEDİĞİMİZ SUTUNLARI YAZIYORUZ

billboard.melt(
    id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
    var_name='week',
    value_name='rating'
).head()
```

```python
# groupby ile AYIRIP ORTALAMA DEĞERLERİ GÖRDÜK
#  DÜZENLİ BİR SYNTAX İÇİN DIKKAT EDEBİLİRSİNİZ.

(billboard
    .melt(
        id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
        var_name='week',
        value_name='rating')
    .groupby("artist")["rating"]
    .mean()
)
```

```python
# groupby ile AYIRIP ORTALAMA DEĞERLERİ GÖRDÜK
#  BURADA TERS SLASH \ KULLANILMIŞTIR PYTHON'A OZGU BİR OZELLIK
(billboard\
    .melt(
        id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
        var_name='week',
        value_name='rating')\
    .groupby("artist")["rating"]\
    .mean()
)
```

```python
# melt()
# id_vars = İSTEDİĞİMİZ SUTUNLARI YAZIYORUZ

billboard_long = pd.melt(
    billboard,
    id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
    var_name='week',
    value_name='rating'
)
```

```python
billboard_long.head()
```

```python
billboard_long.shape
```

```python
# DİĞER VERİ SETİMİZ ZAMAN SERİSİ ANALİZİ İÇİN

ebola = pd.read_csv('../data/country_timeseries.csv')
```

```python
# VERİ SETİMİZİN SON 5 SATIRI tail() İLE GORUNTULENIR
ebola.tail()
```

```python

ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
ebola_long.head()
```

```python
# var_name VE value_name DEFAULT OLARAK VERİLDİĞİNE DİKKAT

ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'],
                    var_name="cd_country",value_name="count")
ebola_long.head()
```

```python
# PYTHON STRİNG OZELLIKLERINDEN split() METHODU İLE AYIRMA

'Cases_Guinea'.split('_')
```

```python
# string İŞLEMİ cd_country SUTUNUNU AYIRMA

variable_split = ebola_long['cd_country'].str.split('_')
variable_split.head()
```

```python
# TUM SUTUN DEĞERLERİ ALINIR

variable_split.str.get(0).head()
```

```python
# YENİ BİR SUTUNA AKTARMAK ICIN

status_values = variable_split.str.get(0)
status_values.head()
```

```python
# AYNI SEKILDE IKINCI BİR SUTUN OLUSTURMAK ICIN

country_values = variable_split.str.get(1)
country_values.head()
```

```python
# YENİ SUTUNLARI OLUŞTURUP İSİMLENDİRDİK

ebola_long['status'] = status_values
ebola_long['country'] = country_values
```

```python
ebola_long.head()
```

```python
#  SUTUNLARA AYIRMA ISLEMI BOYLE BIR SECENEK ILE DE OLABILIRDI

variable_split = ebola_long['cd_country'].str.split('_', expand=True)
variable_split.head()
```

```python
type(variable_split)
```

```python
# variable_split İSİMLİ DATAFRAME'IN SUTUNLARINI ebola_long'a ATAMA YAPIYORUZ

ebola_long[['case2', 'country2']] = variable_split
```

```python
ebola_long.head()
```

```python
variable_split.head()
```

```python
# concat METHODU İLE DE BİRLESTİRME İŞLEMI YAPILABİLİR
# 2 AYRI DATA FRAME'MİMİZİ axis=1 DİYEREK BİRLEŞTİRDİK

ebola_long = pd.concat([ebola_long, variable_split], axis=1)
```
